```python
import pandas as pd
```

```python
ebbu_iscx = pd.read_csv("processed/ebbu&iscx_5.csv")
phishtank = pd.read_csv("processed/phishtank_df_3.csv")
legit_df = pd.read_csv("processed/ligit_df_3.csv")
```

```python
phishing_df = pd.concat([ebbu_iscx, phishtank]).reset_index(drop=True)
print(len(phishing_df), " rows before dropping duplicates")
phishing_df = phishing_df.drop_duplicates(subset="link", keep="last")
print(len(phishing_df), " rows after dropping duplicates")
phishing_df
```

```python
def limpieza(df, archivo):
    # etiqueta los links que tienen ip y limpia los protocolos
    
    # etiquetar si tiene un ip o no en la columna 'has ip'
    df["has ip"] = df["ip_address"].notna()
    df.drop('ip_address', inplace=True, axis=1)
    
    # quitar los que no tienen protocolo
    df = df[df["protocolo"].notna()]
    
    # sacar el '://' del protocolo
    df["protocolo"] = df["protocolo"].squeeze().apply(lambda x: str(x).replace("://",""))
    dfip = df[df['has ip'] == True]
    
    print("se guardo la bd sin IPs como ", archivo+".csv")
    df.to_csv(archivo+".csv", index = False)
    
    print("se guardo la bd con IPs como ", archivo+"_ip.csv")
    dfip.to_csv(archivo+"_ip.csv", index = False)
    
    return df,dfip
```

```python
phishing_cl, phishing_ip = limpieza(phishing_df,'processed/phishing_df')
legit_cl, legit_ip = limpieza(legit_df,'processed/legit_df')
```
